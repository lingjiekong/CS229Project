```python
# import library
import pandas as pd
import numpy as np
```

```python
#------------------------------------------------------ prepare dict -------------------------------------------------------
```

```python
# load the data
trainLoad = pd.read_csv("clicks_train.csv") # split later for train and test
# trim the train and test data first
data = trainLoad[0:1000002] # display_id up to 195916
```

```python
# make dictionary from display_id to ad_id and click tuple
displayid = 1
dataDict = dict()
eachDict = []
totalLen = len(data)
for i in range(totalLen):
    if (data.loc[i]["display_id"] == displayid):
        eachDict.append((data.loc[i]["ad_id"], data.loc[i]["clicked"]))
    else:
        dataDict[displayid] = eachDict  
        displayid += 1
        eachDict = list()
        eachDict.append((data.loc[i]["ad_id"], data.loc[i]["clicked"]))
```

```python
# load event
event = pd.read_csv("events.csv")
# trim data based on last display_id from data
lastDisplayid = 195916
eventParital = eventPartial = event[0:lastDisplayid+1]
```

```python
# make dictionary from display_id to document_id
dis2Doc = dict()
for i in range(len(eventParital)):
    dis2Doc[i] = eventPartial.loc[i]["document_id"]
```

```python
# make dictionary from document_id to display_id by flipping the above value
doc2Dis = dict()
for k, v in dis2Doc.items():
    doc2Dis[v] = k
```

```python
#------------------------------------------------------ start of document -------------------------------------------------------
```

```python
# load document_catetories data and start trim
docCategories = pd.read_csv("documents_categories.csv")
docCategoriesPartial = docCategories[0:1000001] # load first 1000001
```

```python
# build dictionary from document_id to category_id filtering by 
# high confidence level
lenDocCategories = len(docCategoriesPartial)
docid = docCategories.loc[0]["document_id"]
lastDocid = 0
maxConfi = -float('inf')
catogory = 0
doc2cate = dict()
for i in range(lenDocCategories):
    if (docCategories.loc[i]["document_id"] == docid):
        Confi = docCategories.loc[i]["confidence_level"]
        if Confi > maxConfi:
            maxConfi = Confi
            catogory= docCategories.loc[i]["category_id"]
        lastDocid = docid
    else:
        # update the dictionary
        doc2cate[lastDocid] = catogory
        # for the next one
        docid = docCategories.loc[i]["document_id"]
        maxConfi = docCategories.loc[i]["confidence_level"]
        catogory = docCategories.loc[i]["category_id"]
```

```python
# filter the original data the first time based on appearance of document_id 
# in the doucments_categories file
# use doc2cate, doc2Dis, dataDict
postDataDict = dict()
postDoc2Dis = dict()
postDoc2Cate = dict()
for key in doc2cate:
    if key in doc2Dis.keys():
        postDoc2Cate[key] = doc2cate[key]
        postDoc2Dis[key] = doc2Dis[key]
        dis = doc2Dis[key]
        postDataDict[dis] = dataDict[dis]
```

```python
# check our how many percentage of post is left compared to the original one
print ((len(postDataDict))/(len(dataDict)))
print (len(postDataDict))
# 3% left and 6466
```

```python
# load document_topics data and start trim
docTopics = pd.read_csv("documents_topics.csv")
docTopicsPartial = docTopics[0:1000015] # load first 1000015
```

```python
# build dictionary from document_id to topics_id filtering by 
# high confidence level
lenDocTopics = len(docTopicsPartial)
docid = docTopics.loc[0]["document_id"]
lastDocid = 0
maxConfi = -float('inf')
topic = 0
doc2Topic = dict()
for i in range(lenDocTopics):
    if (docTopics.loc[i]["document_id"] == docid):
        Confi = docTopics.loc[i]["confidence_level"]
        if Confi > maxConfi:
            maxConfi = Confi
            topic = docTopics.loc[i]["topic_id"]
        lastDocid = docid
    else:
        # update the dictionary
        doc2Topic[lastDocid] = topic
        # for the next one
        docid = docTopics.loc[i]["document_id"]
        maxConfi = docTopics.loc[i]["confidence_level"]
        topic = docTopics.loc[i]["topic_id"]
```

```python
# filter the original data the second time based on appearance of document_id 
# in the doucments_topics file
# use postDoc2Cate, postDoc2Dis, postDataDict, and doc2Topics
postPostDataDict = dict()
postPostDoc2Dis = dict()
postPostDoc2Cate = dict()
postPostDoc2Topic = dict()
for key in doc2Topic:
    if key in postDoc2Dis.keys():
        postPostDoc2Cate[key] = postDoc2Cate[key]
        postPostDoc2Dis[key] = postDoc2Dis[key]
        postPostDoc2Topic[key] = doc2Topic[key]
        dis = postDoc2Dis[key]
        postPostDataDict[dis] = postDataDict[dis]
```

```python
# check our how many percentage of post is left compared to the last filtered data one
print ((len(postPostDataDict))/(len(postDataDict)))
print (len(postPostDataDict))
# 50% left and 3273
```

```python
# load document_meta data and start trim
docMeta = pd.read_csv("documents_meta.csv")
docMetaPartial = docMeta[0:1000000] # load first 1000000
```

```python
# build dictionary from document_id to topics_id filtering by 
# high confidence level
lenDocMeta = len(docMetaPartial)
lastDocid = 0
doc2Meta = dict()
for i in range(lenDocMeta):
    docid = docMeta.loc[i]["document_id"]
    doc2Meta[docid] = docMeta.loc[i]["source_id"]
```

```python
# filter the original data the third time based on appearance of document_id 
# in the doucments_meta file
# use postDoc2Cate, postDoc2Dis, postDataDict, doc2Topics, and 
postPostPostDataDict = dict()
postPostPostDoc2Dis = dict()
postPostPostDoc2Cate = dict()
postPostPostDoc2Topic = dict()
postPostPostDoc2Meta = dict()
for key in doc2Meta:
    if key in postPostDoc2Dis.keys():
        postPostPostDoc2Cate[key] = postPostDoc2Cate[key]
        postPostPostDoc2Dis[key] = postPostDoc2Dis[key]
        postPostPostDoc2Topic[key] = postPostDoc2Topic[key]
        postPostPostDoc2Meta[key] = doc2Meta[key]
        dis = postDoc2Dis[key]
        postPostPostDataDict[dis] = postPostDataDict[dis]
```

```python
# check our how many percentage of post is left compared to the last filtered data one
print ((len(postPostPostDataDict))/(len(postPostDataDict)))
print (len(postPostPostDataDict))
# 100% left and 3273
```

```python
#----------------------------------------------------- start of ad ----------------------------------------------------------
```

```python
# load promotendContent which contains advertiser_id and campaign_id for ad_id
promotedContent = pd.read_csv("promoted_content.csv")
```

```python
totalLen = len(promotedContent)
ad2Adv = dict()
ad2Cam = dict()
for i in range(totalLen):
    adID = promotedContent.loc[i]["ad_id"]
    advID = promotedContent.loc[i]["advertiser_id"]
    camID = promotedContent.loc[i]["campaign_id"]
    # feature
    ad2Adv[adID] = advID
    ad2Cam[adID] = camID
```

```python
#------------------------------------------------- Start of big dict for NN ----------------------------------------------------
# rename postPostPost to final
# first mapping
finalDoc2Dis = postPostPostDoc2Dis

# flip first mapping
finalDis2Doc = dict()
for k, v in finalDoc2Dis.items():
    finalDis2Doc[v] = k
    
# feature
finalDoc2Cate = postPostPostDoc2Cate
finalDoc2Topic = postPostPostDoc2Topic
finalDoc2Meta = postPostPostDoc2Meta

# data
finalData = postPostPostDataDict

# input data
# finalData 
    # dictionary from display_id to (ad_id, clicked) tuple

# 5 features
# now we have the below dictionary
# ad_id -> avertiser_id/campaign_id
    # ad2Adv (from ad_id to advertiser_id)
    # ad2Cam ( from ad_id) to campaign_id)
    
# display_id -> document_id -> category_id/Meta_id(source_id)/topic_id
    # finalDis2Doc (from display_id to document_id) -> finalDoc2Cate (from document_id to category_id)
    # finalDis2Doc (from display_id to document_id) -> finalDoc2Topic (from document_id to topic_id)
    # finalDis2Doc (from display_id to document_id) -> finalDoc2Meta (from document_id to meta_id(source_id))

combineDataTrainX = dict()
combineDataTrainY = dict()
combineDataTestX = dict()
combineDataTestY = dict()
countSplit = 0
splitThreshold = 2800
for disID in finalData:
    adClickIDTupleSet = finalData[disID]
    for adClickIDTuple in adClickIDTupleSet:
        adID = adClickIDTuple[0]
        clickID = adClickIDTuple[1]
        # from ad_id
        advID = ad2Adv[adID]
        camID = ad2Cam[adID]
        # docID key
        docID = finalDis2Doc[disID]
        # from display_id
        cateID = finalDoc2Cate[docID]
        topicID = finalDoc2Topic[docID]
        metaID = finalDoc2Meta[docID]
        # make new dictionary
        if (countSplit < splitThreshold):
            combineDataTrainX[(disID,adID)] = (countSplit, advID, camID, cateID, topicID, metaID)
            combineDataTrainY[(disID,adID)] = clickID
        else:
            combineDataTestX[(disID,adID)] = (countSplit, advID, camID, cateID, topicID, metaID)
            combineDataTestY[(disID,adID)] = clickID
    countSplit += 1
```

```python
# convert to string, so it can be saved by using json
ad2AdvStr = dict()
for each in ad2Adv:
    ad2AdvStr[str(each)] = str(ad2Adv[each])

ad2CamStr = dict()
for each in ad2Cam:
    ad2CamStr[str(each)] = str(ad2Cam[each])
    
finalDis2DocStr = dict()
for each in finalDis2Doc:
    finalDis2DocStr[str(each)] = str(finalDis2Doc[each])
    
finalDoc2CateStr = dict()
for each in finalDoc2Cate:
    finalDoc2CateStr[str(each)] = str(finalDoc2Cate[each])
    
finalDoc2TopicStr = dict()
for each in finalDoc2Topic:
    finalDoc2TopicStr[str(each)] = str(finalDoc2Topic[each])
    
finalDoc2MetaStr = dict()
for each in finalDoc2Meta:
    finalDoc2MetaStr[str(each)] = str(finalDoc2Meta[each])
    
finalDataStr = dict()
for each in finalData:
    finalDataStr[str(each)] = repr(finalData[each])
```

```python
import json
# each separate dict()
with open('ad2Adv.json', 'w') as fp:
    json.dump(ad2AdvStr, fp, sort_keys=True, indent=4)

with open('ad2Cam.json', 'w') as fp:
    json.dump(ad2CamStr, fp, sort_keys=True, indent=4)

with open('finalDis2Doc.json', 'w') as fp:
    json.dump(finalDis2DocStr, fp, sort_keys=True, indent=4) 
    
with open('finalDoc2Cate.json', 'w') as fp:
    json.dump(finalDoc2CateStr, fp, sort_keys=True, indent=4) 

with open('finalDoc2Topic.json', 'w') as fp:
    json.dump(finalDoc2TopicStr, fp, sort_keys=True, indent=4) 
    
with open('finalDoc2Meta.json', 'w') as fp:
    json.dump(finalDoc2MetaStr, fp, sort_keys=True, indent=4) 
    
with open('finalData.json', 'w') as fp:
    json.dump(finalDataStr, fp, sort_keys=True, indent=4) 


# # train and test dict()
# with open('combineDataTrainX.json', 'w') as fp:
#     json.dump(combineDataTrainX, fp, sort_keys=True, indent=4)
    
# with open('combineDataTrainY.json', 'w') as fp:
#     json.dump(combineDataTrainX, fp, sort_keys=True, indent=4)
    
# with open('combineDataTestX.json', 'w') as fp:
#     json.dump(combineDataTrainX, fp, sort_keys=True, indent=4)
    
# with open('combineDataTestY.json', 'w') as fp:
#     json.dump(combineDataTrainX, fp, sort_keys=True, indent=4)
```
